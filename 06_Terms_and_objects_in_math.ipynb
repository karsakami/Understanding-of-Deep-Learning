#Python implementation in numpy

```python
import numpy as np
import torch
```

Create vector

```python
v = np.array([ [1,0,2,5,-2] ])
print(v)
```

Transpose it

```python
print(v.T)
```

Create and transpose matrix

```python
m = np.array([ [8,5,2,4,2],
              [8,5,7,4,2] ])
```

```python
print(m)
```

```python
print(m.T)
```
