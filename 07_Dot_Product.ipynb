## Lets do this in numpy 
Difference about them is that a numpy dot product is just a number, but torch dot product is a tensor. 

```python
import numpy as np
import torch
```

```python
# create a vector
nv1 = np.array([1,2,3,4])
nv2 = np.array([2,7,5,3])

# dot product via numpy function
np.dot(nv1,nv2)
```

## Lets do this in pytorch 

```python
tv1 = torch.tensor([1,2,3,4])
tv2 = torch.tensor([1,2,3,4])

torch.dot(tv1,tv2)
```
