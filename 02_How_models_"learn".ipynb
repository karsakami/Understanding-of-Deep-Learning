## **We will now learn**


*  Summary into learning mechanism of artificial neural networks **(ANN)**
*  Difference between **forward** and **backwards propagation**
*  Kinda analogy for understanding how deep networks learn


###Lets assume that you are a chef at restaurant, and you want to improve your dishes. 

So what you do to get better? You ask customers what they think about your dishes. 

One person says 'its too sweet' 

So what ANN model will do? He will produce more and more dishes ($y$), and learn from mistakes, until the customer is happy, algorithm will adjust the weight of the features (blue,  $x_1, x_2$..), such as salt, sugar, pepper etc.

image.png

####Our model look more like this in this example, so we have $x_0$ - the bias, value with no weight, because this value stays the same all the time. 

####The features of our sandwich, such as butter and cheese. We combine them in the right way and we get sandwich.

#### So basically the neural network learns by **two steps**:


*   **Forward propagation** - which is just making sandwich
*   **Backwards propagation** - which is getting message what is wrong, and then model try to fix weights to get better result in next forward propagation, and it repeats and repeats

We can remember that **forward prop** goes from the input to the right, and backwards is just opposite.
