# **Implementation**
###We multiply matrices by "@" eg. (M1@M2)

```python
import numpy as np
```

```python
# create random matrices
A = np.random.randn(3,4)
B = np.random.randn(4,5)
C = np.random.randn(3,7)

np.round(A@B, 0)
```

## We cant multiply matrix $C$ with $A$, but if we transpose it we can, because the number of rows in first is same as number of rows in second 

```python
np.round(C.T@A, 0)
```
