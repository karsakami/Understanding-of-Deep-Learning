### **Philosophy of deep learning**


*   Lots of small, simple things is better than one complex thing (We are talking about deep learning architecture right now - big amount of simple equations, repeated)
*   DL is fundamentally empirical. What does it means? Instead of traditional programming - giving a instructions to computer, we build a model, and we give the model a lots of data, lots of examples and lots of feedback. \
So we can say that we are putting trust into technology that we dont understand. Same as we trust a car and phone, I know how to drive a car, but I dont know how it works.
